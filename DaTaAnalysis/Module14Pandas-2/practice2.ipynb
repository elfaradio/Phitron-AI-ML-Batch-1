```python
import pandas as pd
```

```python
def count_salary_categories(accounts: pd.DataFrame) -> pd.DataFrame:

    obj = {
        "Low Salary": 0,
        "Average Salary": 0,
        "High Salary": 0
    }

    obj["Low Salary"] = (accounts['income'] < 20000).sum()
    obj["Average Salary"] = (
        (accounts['income'] >= 20000) & (accounts['income'] <= 50000)).sum()

    obj["High Salary"] = (accounts['income'] > 50000).sum()

    df = pd.DataFrame({
        'category': obj.keys(),
        'accounts_count': obj.values()
    })

    return df
```
