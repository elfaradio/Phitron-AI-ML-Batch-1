```python
import pandas as pd
import numpy as np


df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})

print(df, "\n")

ans1 = df.fillna({'A': 0, 'B': 99, 'C': 'Unknown'})
print(ans1, "\n")

ans2 = df.copy()
ans2['B'] = ans2['B'].ffill()
print(ans2, "\n")


def solve(df: pd.DataFrame) -> pd.DataFrame:
    x = df.copy()
    for c in x.columns:
        if x[c].dtype in ['float64', 'int64']:
            x[c].fillna(x[c].mean(), inplace=True)
        else:
            x[c].fillna('Unknown', inplace=True)
    return x


ans3 = solve(df)
print(ans3)
```
