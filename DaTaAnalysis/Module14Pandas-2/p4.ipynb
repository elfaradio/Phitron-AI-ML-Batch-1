```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})

print(df)

# Drop rows where column A has missing values

ans1 = df.dropna(subset=['A'])
print(ans1, "\n")

# Drop rows with less than 2 non-null values

ans2 = df.dropna(thresh=2)
print(ans2, "\n")

# Drop rows where both Column A and Column B has missing values

ans3 = df.dropna(subset=['A', 'B'], how='all')
print(ans3, "\n")
```
