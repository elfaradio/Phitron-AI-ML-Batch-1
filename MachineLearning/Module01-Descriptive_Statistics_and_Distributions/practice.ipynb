Dataset: The annual salaries (in $1000) of 20 employees in a tech company:
[45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500]

Calculate:
Range
Variance
Standard deviation
Q2
IQR (Interquartile Range)
Calculate the lower and upper fences for outlier detection
Identify any outliers using the IQR method
How many outliers are there and what are their values?


```python
import numpy as np
import pandas as pd
```

```python
data = [45, 48, 50, 52, 55, 58, 60, 62, 65, 68,
        70, 72, 75, 78, 80, 85, 90, 95, 150, 500]
arr = np.array(data)
print(arr.max()-arr.min())  # Range
```

```python
var1 = arr.var(ddof=0)
var2 = arr.var(ddof=1)
print(var1, var2)
```

```python
std1 = arr.std(ddof=0)
std2 = arr.std(ddof=1)
print(std1, std2)
```

```python
q2 = np.percentile(arr, 50)
print(q2)
```

```python
iqr = (np.percentile(arr, 75)-np.percentile(arr, 25))
print(iqr)
```

```python
lower_fences = np.percentile(arr, 25)-(1.5*iqr)
higher_fences = np.percentile(arr, 75)+(1.5*iqr)
print(lower_fences, higher_fences)
```

```python
outliers = arr[(arr < lower_fences) | (arr > higher_fences)]
print(outliers)
print(len(outliers))
```
