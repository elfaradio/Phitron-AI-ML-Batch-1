Part A – Central Tendency and Spread

1. Marks: 50, 60, 65, 70, 75, 80, 85, 90
   Find mean, median, mode.
   Replace 90 with 900 and recompute mean. Explain the effect of outliers.
2. Temperatures (°C): 29, 31, 33, 33, 32, 31, 30
   Calculate variance and standard deviation.
   Which measure is easier to interpret and why?


```python
import numpy as np
import pandas as pd
```

```python
arr = np.array([50, 60, 65, 70, 75, 75, 80, 85, 90])
print(np.mean(arr))
print(np.median(arr))
# print(arr.mode())
df = pd.DataFrame(data=arr, columns=['val'])
print(df['val'].mode())
```

```python
arr[8] = 900
print(np.mean(arr))
print(np.median(arr))
# print(arr.mode())
df1 = pd.DataFrame(data=arr, columns=['val'])
print(df1['val'].mode())
```

```python
tt = np.array([29, 31, 33, 33, 32, 31, 30])

variance = np.var(tt, ddof=0)
std_dev = np.std(tt, ddof=0)

print("Variance:", variance)
print("Standard Deviation:", std_dev)
```

Part B – Percentiles, IQR and Z-Score 3. Cat weights (kg): 2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5
Find P25, P50, P75 and IQR. Mark possible outliers. 4. Given μ = 50, σ = 10
Find Z for x = 65 and interpret.
For Z = −2.5, find x.


```python
w = np.array([2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5])

p25 = np.percentile(w, 25)
p50 = np.percentile(w, 50)
p75 = np.percentile(w, 75)
iqr = p75 - p25

lb = p25 - 1.5 * iqr
ub = p75 + 1.5 * iqr

ans = w[(w < lb) | (w > ub)]

print(p25, p50, p75, iqr, lb, ub)
print(ans)  # No outliers
```

```python
x = 65
mu = 50
sigma = 10
z_score = (x-mu)/sigma
print(z_score)
```

```python
z = -2.5
x = (z*sigma)+mu
print(x)
```
